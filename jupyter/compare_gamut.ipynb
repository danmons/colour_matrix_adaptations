# Compare D65 and D93 gamuts

Scroll down for rendered outputs.

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection
import numpy as np

```

Set up our numbers.  Values copy/pasted lazily from the CSV files in this repo.

```python
plt.rcParams["figure.figsize"] = (10,10)

# Colourspace xyY values

BT709_Wx = 0.3127000000000000
BT709_Wy = 0.3290000000000000

BT709_Rx = 0.6400000000000000
BT709_Ry = 0.3300000000000000
BT709_Gx = 0.3000000000000000
BT709_Gy = 0.6000000000000000
BT709_Bx = 0.1500000000000000
BT709_By = 0.0600000000000000

D93_Wx = 0.2831109374591642
D93_Wy = 0.2970729817807810

BT2020_Rx = 0.7080000000000000
BT2020_Ry = 0.2920000000000000
BT2020_Gx = 0.1700000000000000
BT2020_Gy = 0.7970000000000000
BT2020_Bx = 0.1310000000000000
BT2020_By = 0.0460000000000000

# vertices of a pyramid.
# R/G/B Y values are always 0 on the base.
# Peak is 1 for SDR, 4 for HDR400, and uniquely scaled for special versions

v_BT709_D65 = np.array( [[BT709_Rx, BT709_Ry, 0],
                         [BT709_Gx, BT709_Gy, 0],
                         [BT709_Bx, BT709_By, 0],
                         [BT709_Wx, BT709_Wx, 1]])

v_BT709_D93_100 = np.array( [[BT709_Rx, BT709_Ry, 0],
                             [BT709_Gx, BT709_Gy, 0],
                             [BT709_Bx, BT709_By, 0],
                             [D93_Wx, D93_Wy, 1]])

v_BT2020_400 = np.array( [[BT2020_Rx, BT2020_Ry, 0],
                          [BT2020_Gx, BT2020_Gy, 0],
                          [BT2020_Bx, BT2020_By, 0],
                          [BT709_Wx, BT709_Wy, 4]])

v_BT2020_120 = np.array( [[BT2020_Rx, BT2020_Ry, 0],
                          [BT2020_Gx, BT2020_Gy, 0],
                          [BT2020_Bx, BT2020_By, 0],
                          [BT709_Wx, BT709_Wy, 1.2]])

v_BT709_D93_090 = np.array( [[BT709_Rx, BT709_Ry, 0],
                             [BT709_Gx, BT709_Gy, 0],
                             [BT709_Bx, BT709_By, 0],
                             [D93_Wx, D93_Wy, 0.955]])

v_BT709_D93_080 = np.array( [[BT709_Rx, BT709_Ry, 0],
                             [BT709_Gx, BT709_Gy, 0],
                             [BT709_Bx, BT709_By, 0],
                             [D93_Wx, D93_Wy, 0.905]])

v_BT709_D93_074 = np.array( [[BT709_Rx, BT709_Ry, 0],
                             [BT709_Gx, BT709_Gy, 0],
                             [BT709_Bx, BT709_By, 0],
                             [D93_Wx, D93_Wy, 0.875]])

# scaling notes
# 74 percent clip point
#0.875	0.745448622658488
#
# 80 percent - scale 0.91
#0.905  0.802836066149821
#
# 90 percent 
# 0.955  0.903664907634661


# generate list of sides' polygons of our pyramid
p_BT709_D65 = [[v_BT709_D65[2], v_BT709_D65[1], v_BT709_D65[3]],
               [v_BT709_D65[1], v_BT709_D65[0], v_BT709_D65[3]],
               [v_BT709_D65[2], v_BT709_D65[0], v_BT709_D65[3]],
               [v_BT709_D65[2], v_BT709_D65[1], v_BT709_D65[0]],
               ]

p_BT709_D93_100 = [[v_BT709_D93_100[2], v_BT709_D93_100[1], v_BT709_D93_100[3]],
                   [v_BT709_D93_100[1], v_BT709_D93_100[0], v_BT709_D93_100[3]],
                   [v_BT709_D93_100[2], v_BT709_D93_100[0], v_BT709_D93_100[3]],
                   [v_BT709_D93_100[2], v_BT709_D93_100[1], v_BT709_D93_100[0]],
                   ]

p_BT2020_400 = [[v_BT2020_400[2], v_BT2020_400[1], v_BT2020_400[3]],
                [v_BT2020_400[1], v_BT2020_400[0], v_BT2020_400[3]],
                [v_BT2020_400[2], v_BT2020_400[0], v_BT2020_400[3]],
                [v_BT2020_400[2], v_BT2020_400[1], v_BT2020_400[0]],
                ]

p_BT2020_120 = [[v_BT2020_120[2], v_BT2020_120[1], v_BT2020_120[3]],
                [v_BT2020_120[1], v_BT2020_120[0], v_BT2020_120[3]],
                [v_BT2020_120[2], v_BT2020_120[0], v_BT2020_120[3]],
                [v_BT2020_120[2], v_BT2020_120[1], v_BT2020_120[0]],
                ]

p_BT709_D93_090 = [[v_BT709_D93_090[2], v_BT709_D93_090[1], v_BT709_D93_090[3]],
               [v_BT709_D93_090[1], v_BT709_D93_090[0], v_BT709_D93_090[3]],
               [v_BT709_D93_090[2], v_BT709_D93_090[0], v_BT709_D93_090[3]],
               [v_BT709_D93_090[2], v_BT709_D93_090[1], v_BT709_D93_090[0]],
               ]

p_BT709_D93_080 = [[v_BT709_D93_080[2], v_BT709_D93_080[1], v_BT709_D93_080[3]],
               [v_BT709_D93_080[1], v_BT709_D93_080[0], v_BT709_D93_080[3]],
               [v_BT709_D93_080[2], v_BT709_D93_080[0], v_BT709_D93_080[3]],
               [v_BT709_D93_080[2], v_BT709_D93_080[1], v_BT709_D93_080[0]],
               ]

p_BT709_D93_074 = [[v_BT709_D93_074[2], v_BT709_D93_074[1], v_BT709_D93_074[3]],
                   [v_BT709_D93_074[1], v_BT709_D93_074[0], v_BT709_D93_074[3]],
                   [v_BT709_D93_074[2], v_BT709_D93_074[0], v_BT709_D93_074[3]],
                   [v_BT709_D93_074[2], v_BT709_D93_074[1], v_BT709_D93_074[0]],
                   ]
```

# D65
Let's draw a D65 colour volume.  Note that colour volume doesn't look like this, but we're simplifying this greatly for the sake of explanation.  

The base of this pyramid are the co-ordinates for the Red, Green and Blue primaries. The top is the D65 white point, where all three colours converge as they are added together. 

We'll colour this red for visual clarity.

```python
fig_d65 = plt.figure()
a_d65 = fig_d65.add_subplot(111, projection='3d')

a_d65.axes.set_xlim3d(left=0.0, right=1.0) 
a_d65.axes.set_ylim3d(bottom=0.0, top=1.0) 
a_d65.axes.set_zlim3d(bottom=0.0, top=1.0) 

# plot sides
a_d65.add_collection3d(Poly3DCollection(p_BT709_D65, 
 facecolors='red', linewidths=1, edgecolors='black', alpha=.2))


plt.draw()
```

# D93

Now let's draw our D93 colour volume.  Again, the base are the co-ordinates for Red, Green and Blue primaries.  The top of the pyramid is now the D93 white point.  We'll colour this blue for visual clarity.


```python
fig_d93 = plt.figure()
a_d93 = fig_d93.add_subplot(111, projection='3d')

a_d93.axes.set_xlim3d(left=0.0, right=1.0) 
a_d93.axes.set_ylim3d(bottom=0.0, top=1.0) 
a_d93.axes.set_zlim3d(bottom=0.0, top=1.0) 

# plot sides

a_d93.add_collection3d(Poly3DCollection(p_BT709_D93_100, 
 facecolors='blue', linewidths=1, edgecolors='black', alpha=.2))

plt.draw()
```

# D65 vs D93
Let's overlay D65 and D93.

Everything in red is D65 only.  Everything in blue is D93 only. 

Where they overlap and share colour volume, it will look purple. 

We see a problem here, in that if our screen is D65, there are going to be some colours in D93 we simply cannot show.  These colours are going to be clipped out. 

```python
fig_d65_d93 = plt.figure()
a_d65_d93 = fig_d65_d93.add_subplot(111, projection='3d')

a_d65_d93.axes.set_xlim3d(left=0.0, right=1.0) 
a_d65_d93.axes.set_ylim3d(bottom=0.0, top=1.0) 
a_d65_d93.axes.set_zlim3d(bottom=0.0, top=1.0) 

# plot sides
a_d65_d93.add_collection3d(Poly3DCollection(p_BT709_D65, 
 facecolors='red', linewidths=1, edgecolors='black', alpha=.2))

a_d65_d93.add_collection3d(Poly3DCollection(p_BT709_D93_100, 
 facecolors='blue', linewidths=1, edgecolors='black', alpha=.2))

plt.draw()
```

# D65 vs D93 scaled to 90%

Let's scale back the D93 value to 90% brightness.  We start to see less clipping at the expense of peak brightness.

```python
fig_d65_d93_090 = plt.figure()
a_d65_d93_090 = fig_d65_d93_090.add_subplot(111, projection='3d')

a_d65_d93_090.axes.set_xlim3d(left=0.0, right=1.0) 
a_d65_d93_090.axes.set_ylim3d(bottom=0.0, top=1.0) 
a_d65_d93_090.axes.set_zlim3d(bottom=0.0, top=1.0) 

# plot sides
a_d65_d93_090.add_collection3d(Poly3DCollection(p_BT709_D65, 
 facecolors='red', linewidths=1, edgecolors='black', alpha=.2))

a_d65_d93_090.add_collection3d(Poly3DCollection(p_BT709_D93_090, 
 facecolors='blue', linewidths=1, edgecolors='black', alpha=.2))

plt.draw()
```

# D65 vs D93 scaled to 80%

Let's scale back D93 even further, to 80%.  Less clipping again. 

```python
fig_d65_d93_080 = plt.figure()
a_d65_d93_080 = fig_d65_d93_080.add_subplot(111, projection='3d')

a_d65_d93_080.axes.set_xlim3d(left=0.0, right=1.0) 
a_d65_d93_080.axes.set_ylim3d(bottom=0.0, top=1.0) 
a_d65_d93_080.axes.set_zlim3d(bottom=0.0, top=1.0) 


# plot sides
a_d65_d93_080.add_collection3d(Poly3DCollection(p_BT709_D65, 
 facecolors='red', linewidths=1, edgecolors='black', alpha=.2))

a_d65_d93_080.add_collection3d(Poly3DCollection(p_BT709_D93_080, 
 facecolors='blue', linewidths=1, edgecolors='black', alpha=.2))

plt.draw()
```

# D65 vs D93 scaled to 74%

At approximately 74%, all clipping of the simulated D93 is gone, and the simulated volume fits entirely inside our D65 colour volume.  

We can achieve an entire simulated D93 colour volume for around a 25% (pre-gamma) brightness sacrifice. 

```python
fig_d65_d93_074 = plt.figure()
a_d65_d93_074 = fig_d65_d93_074.add_subplot(111, projection='3d')

a_d65_d93_074.axes.set_xlim3d(left=0.0, right=1.0) 
a_d65_d93_074.axes.set_ylim3d(bottom=0.0, top=1.0) 
a_d65_d93_074.axes.set_zlim3d(bottom=0.0, top=1.0) 

# plot sides
a_d65_d93_074.add_collection3d(Poly3DCollection(p_BT709_D65, 
 facecolors='red', linewidths=1, edgecolors='black', alpha=.2))

a_d65_d93_074.add_collection3d(Poly3DCollection(p_BT709_D93_074, 
 facecolors='blue', linewidths=1, edgecolors='black', alpha=.2))

plt.draw()
```

# D65 vs D93 vs BT.2020 HDR 400 nits

Let's compare to BT.2020 wide gamut. It can be both SDR and HDR, but we're going to use the HDR version. Let's set the peak brightness to the lowest HDR standard, "DisplayHDR400", and compare to a full D65 and full D93. 

Even at the lowest HDR standard, BT.2020 has enormous volume, and can totally encapsulate both SDR D65 and SDR D93. 

```python
fig_d65_d93_bt2020 = plt.figure()
a_d65_d93_bt2020 = fig_d65_d93_bt2020.add_subplot(111, projection='3d')

a_d65_d93_bt2020.axes.set_xlim3d(left=0.0, right=1.0) 
a_d65_d93_bt2020.axes.set_ylim3d(bottom=0.0, top=1.0) 
a_d65_d93_bt2020.axes.set_zlim3d(bottom=0.0, top=1.0) 

# plot sides
a_d65_d93_bt2020.add_collection3d(Poly3DCollection(p_BT709_D65, 
 facecolors='red', linewidths=1, edgecolors='black', alpha=.2))

a_d65_d93_bt2020.add_collection3d(Poly3DCollection(p_BT709_D93_100, 
 facecolors='blue', linewidths=1, edgecolors='black', alpha=.2))

a_d65_d93_bt2020.add_collection3d(Poly3DCollection(p_BT2020_400, 
 facecolors='green', linewidths=1, edgecolors='black', alpha=.2))

plt.draw()
```

# D65 vs D93 vs BT.2020 HDR 120 nits

Scaling our BT2020 HDR colourspace right, we see even at a very conservative 120 nits (vs SDR spec 100 nits) we have enough volume to completely encapsulate D65 vs D93. 

```python
fig_d65_d93_bt2020_120 = plt.figure()
a_d65_d93_bt2020_120 = fig_d65_d93_bt2020_120.add_subplot(111, projection='3d')

a_d65_d93_bt2020_120.axes.set_xlim3d(left=0.0, right=1.0) 
a_d65_d93_bt2020_120.axes.set_ylim3d(bottom=0.0, top=1.0) 
a_d65_d93_bt2020_120.axes.set_zlim3d(bottom=0.0, top=1.0) 

# plot sides
a_d65_d93_bt2020_120.add_collection3d(Poly3DCollection(p_BT709_D65, 
 facecolors='red', linewidths=1, edgecolors='black', alpha=.2))

a_d65_d93_bt2020_120.add_collection3d(Poly3DCollection(p_BT709_D93_100, 
 facecolors='blue', linewidths=1, edgecolors='black', alpha=.2))

a_d65_d93_bt2020_120.add_collection3d(Poly3DCollection(p_BT2020_120, 
 facecolors='green', linewidths=1, edgecolors='black', alpha=.2))

plt.draw()
```
